```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
board = np.fromfile("board.dat", dtype=np.int32).reshape((-1,128, 128))
```

```python
fig = plt.figure()
ims = []
for i in range(0,board.shape[0]):
    im = plt.imshow(board[i], animated=True, vmin=0, vmax=1, cmap='Greys')
    ims.append([im])
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat=False)
ani.save("polyom.mp4")
plt.show()
```
