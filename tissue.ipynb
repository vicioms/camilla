```python
import torch
from torch import autograd
import networkx as nx
```

```python
def graph_to_torch(graph: nx.Graph, attribute_name: str):
    nodes = list(graph.nodes())            
    if isinstance(nodes[0], tuple): # node is a tuple, we need to relabel the graph
        mapping = {node: i for i, node in enumerate(nodes)}
        graph = nx.relabel_nodes(graph, mapping)
        nodes = list(graph.nodes())
    attributes = nx.get_node_attributes(graph, attribute_name)
    edge_list = torch.tensor(list(graph.edges()))
    
```

```python
graph_to_torch(nx.hexagonal_lattice_graph(10, 10, periodic=True, with_positions=True, create_using=nx.DiGraph), 'pos')
```
