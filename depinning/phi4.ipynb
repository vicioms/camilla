```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.fromfile("phi4.bin", dtype=np.float32).reshape((-1, 1024,1024))
fig = plt.figure(figsize=(4,4))
plt.plot(data.mean(axis=(1,2)))
plt.show()
```

```python
plt.imshow(data[-500], vmin=-1, vmax=1, cmap='seismic', interpolation=None)
plt.colorbar()
plt.show()
```
